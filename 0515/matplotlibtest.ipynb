```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-np.pi,np.pi, 360, endpoint=True)
type(x)
```

```python
Sin, Cos = np.sin(x), np.cos(x)
```

```python
plt.plot(x, Sin,color='red',linewidth=2.5,linestyle='-',label=r'$cos(t)$')
plt.plot(x, Cos,color='blue',linewidth=2.5,linestyle='-',label=r'$sin(t)$')
plt.legend(loc='upper left',frameon=False)  
```
